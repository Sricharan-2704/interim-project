# 1️⃣ What is Angular & Why Do We Use It?

## What is Angular?

Angular is a **TypeScript-based framework** for building **Single Page Applications (SPAs)**.

### Key Points:
- **Frontend Framework** - Runs in the browser
- **Structured** - Enforces architectural patterns
- **Component-Based** - Break UI into reusable pieces
- **Full-Featured** - Built-in routing, forms, HTTP, etc.
- **TypeScript** - Type-safe JavaScript
- **Reactive** - Uses RxJS for reactive programming

### Traditional Website vs Single Page Application

```
TRADITIONAL WEBSITE:
User clicks link → Browser requests new page → Server sends HTML → Browser reloads
Result: Page blinks, everything reloads

SINGLE PAGE APPLICATION (Angular):
User clicks link → Angular changes view → No page reload → Smooth transition
Result: Fast, smooth, desktop-like experience
```

### Why Angular for Our Sponsorship App?
- **Complex interactions** - Different dashboards for different roles
- **Real-time updates** - Notifications, campaign status
- **Form handling** - Registration, campaign creation, payments
- **Authentication** - Role-based access control
- **Scalability** - Large codebase needs structure
- **Team development** - Everyone uses same patterns

# 4️⃣ CORE ANGULAR CONCEPTS EXPLAINED

## What is a Component?

A component is a **reusable piece of UI** with its own:
- **HTML Template** - How it looks
- **TypeScript Class** - How it behaves/logic
- **CSS Styles** - How it's styled

```
Component = HTML Template + TypeScript Logic + CSS Styles
```

### Example: LoginComponent

**login.component.ts** (TypeScript logic)
```typescript
import { Component } from '@angular/core';
import { AuthService } from '../../services/auth.service';

@Component({
  selector: 'app-login',           // Use component as <app-login/>
  templateUrl: './login.component.html',
  styleUrls: ['./login.component.scss']
})
export class LoginComponent {
  username: string = '';
  password: string = '';
  
  constructor(private authService: AuthService) {}
  
  login() {
    // Business logic here
    this.authService.login({
      email: this.username,
      password: this.password
    }).subscribe(response => {
      console.log('Login successful!');
    });
  }
}
```

**login.component.html** (Template - how it looks)
```html
<div class="login-container">
  <h2>Login</h2>
  <form (ngSubmit)="login()">
    <input placeholder="Email" [(ngModel)]="username" name="username">
    <input placeholder="Password" type="password" [(ngModel)]="password" name="password">
    <button type="submit">Login</button>
  </form>
</div>
```

**login.component.scss** (Styles)
```scss
.login-container {
  padding: 20px;
  border: 1px solid #ccc;
}
```

## Component Lifecycle

Each component goes through a lifecycle:

```
1. Create Component
        ↓
2. constructor()
        ↓
3. ngOnInit() [MOST IMPORTANT]
   - Load initial data
   - Setup subscriptions
   - Call services
        ↓
4. ngOnChange()
   - When @Input properties change
        ↓
5. User interactions
   - Clicks, forms, etc.
        ↓
6. ngOnDestroy()
   - Cleanup
   - Unsubscribe from observables
        ↓
7. Component Destroyed
```

### Common Lifecycle Hooks

```typescript
export class MyComponent implements OnInit, OnDestroy {
  
  ngOnInit(): void {
    // Called once when component is created
    // Perfect place to fetch initial data
    this.loadCampaigns();
  }
  
  ngOnDestroy(): void {
    // Called when component is destroyed
    // Cleanup: unsubscribe from observables
    this.subscription.unsubscribe();
  }
}
```

## What is Data Binding?

Data binding connects TypeScript data to HTML template.

### Types of Data Binding

```
1. PROPERTY BINDING (Component → Template)
   [property]="value"
   <img [src]="imageUrl">
   
2. TWO-WAY BINDING (Both directions)
   [(ngModel)]="value"
   <input [(ngModel)]="username">
   
3. EVENT BINDING (Template → Component)
   (event)="handler()"
   <button (click)="save()">Save</button>
   
4. INTERPOLATION (Component → Template)
   {{ value }}
   <p>Hello {{ userName }}</p>
```

## What is a Service?

A service is a **class that provides reusable functionality** to components.

```
Service = Shared business logic in one place

Benefits:
- Reuse logic across components
- API calls (HttpClient)
- State management (BehaviorSubject)
- Authentication
- etc.
```

### Example: AuthService

```typescript
@Injectable({ providedIn: 'root' })
export class AuthService {
  constructor(private http: HttpClient) {}
  
  login(credentials: LoginRequest): Observable<AuthResponse> {
    return this.http.post('/api/auth/login', credentials);
  }
  
  getToken(): string | null {
    return localStorage.getItem('token');
  }
}

// Multiple components use same service
@Component(...)      // LoginComponent
export class LoginComponent {
  constructor(private authService: AuthService) {}
}

@Component(...)      // DashboardComponent
export class DashboardComponent {
  constructor(private authService: AuthService) {}
}
```

## What is Dependency Injection (DI)?

Dependency Injection is Angular automatically providing ("injecting") instances of services to components.

```
WITHOUT DI (Manual - Bad):
export class LoginComponent {
  authService = new AuthService();  // Creating manually
}

WITH DI (Angular automatic - Good):
export class LoginComponent {
  constructor(private authService: AuthService) {}  // Angular injects it
}
```

### How @Injectable Works

```typescript
@Injectable({
  providedIn: 'root'   // Available app-wide (Singleton)
})
export class AuthService {
  // Single instance shared across entire app
}

// Usage in any component
export class LoginComponent {
  constructor(private authService: AuthService) {}
  // Angular automatically provides the same instance
}
```

# 5️⃣ MODULE SYSTEM - How Angular Organizes Code

## What is an NgModule?

An NgModule is a **container for related code** (components, services, guards, etc.)

```
NgModule = Organized package of:
- Components
- Services
- Pipes
- Guards
- Other modules
```

## AppModule - The Master Module

```typescript
// app.module.ts
@NgModule({
  declarations: [          // Declare components
    AppComponent,
    LoginComponent,
    SignupComponent,
    AdminDashboardComponent,
    // ... all other components
  ],
  imports: [               // Import other modules
    BrowserModule,         // Core Angular features
    HttpClientModule,      // For API calls
    FormsModule,           // For form handling
    ReactiveFormsModule,   // For reactive forms
    AppRoutingModule,      // Routing
    MatToolbarModule,      // Angular Material
    MatButtonModule,
    MatCardModule,
    // ... many more Material modules
  ],
  providers: [             // Provide services
    AuthGuard,
    RoleGuard,
    {
      provide: HTTP_INTERCEPTORS,
      useClass: AuthInterceptor,
      multi: true
    }
  ],
  bootstrap: [AppComponent]  // Root component to load
})
export class AppModule { }
```

## Why Modules?

| Benefit | Without Modules | With Modules |
|---------|-----------------|---------------|
| **Organization** | Everything mixed together | Features organized clearly |
| **Lazy Loading** | All code loaded at startup | Load modules on-demand |
| **Reusability** | Duplicate code | Share modules between apps |
| **Testing** | Hard to isolate | Easy to test modules in isolation |
| **Maintenance** | Confusing | Clear structure |

# 6️⃣ COMPONENTS - THE BUILDING BLOCKS

## Component Anatomy

Every component has 3 files:

### 1. TypeScript File (.ts)
```typescript
import { Component, OnInit } from '@angular/core';
import { AuthService } from '../../services/auth.service';

@Component({
  selector: 'app-login',              // How to use: <app-login></app-login>
  templateUrl: './login.component.html',
  styleUrls: ['./login.component.scss']
})
export class LoginComponent implements OnInit {
  // Properties (data)
  email: string = '';
  password: string = '';
  isLoading: boolean = false;
  errorMessage: string = '';
  
  // Dependency Injection
  constructor(private authService: AuthService) { }
  
  // Lifecycle hook
  ngOnInit() {
    // Initialization code
  }
  
  // Methods (behaviors)
  login() {
    this.isLoading = true;
    this.authService.login({ email: this.email, password: this.password })
      .subscribe({
        next: (response) => {
          console.log('Login successful');
          this.isLoading = false;
        },
        error: (error) => {
          this.errorMessage = 'Invalid credentials';
          this.isLoading = false;
        }
      });
  }
}
```

### 2. HTML Template File (.html)
```html
<div class="login-container">
  <h2>User Login</h2>
  
  <!-- Error message display -->
  <div *ngIf="errorMessage" class="error">
    {{ errorMessage }}
  </div>
  
  <!-- Login form -->
  <form (ngSubmit)="login()">
    <!-- Email input with two-way binding -->
    <input 
      type="email" 
      [(ngModel)]="email" 
      name="email"
      placeholder="Enter email"
      required
    >
    
    <!-- Password input -->
    <input 
      type="password" 
      [(ngModel)]="password" 
      name="password"
      placeholder="Enter password"
      required
    >
    
    <!-- Submit button -->
    <button type="submit" [disabled]="isLoading">
      {{ isLoading ? 'Loading...' : 'Login' }}
    </button>
  </form>
</div>
```

### 3. CSS File (.scss)
```scss
.login-container {
  max-width: 400px;
  margin: 50px auto;
  padding: 20px;
  border: 1px solid #ddd;
  border-radius: 8px;
  
  h2 {
    text-align: center;
    color: #333;
  }
  
  form {
    display: flex;
    flex-direction: column;
    gap: 15px;
    
    input {
      padding: 10px;
      border: 1px solid #ccc;
      border-radius: 4px;
    }
    
    button {
      padding: 10px;
      background-color: #007bff;
      color: white;
      border: none;
      border-radius: 4px;
      cursor: pointer;
    }
  }
  
  .error {
    color: red;
    font-size: 14px;
  }
}
```

## How Services Use HttpClient

```typescript
@Injectable({ providedIn: 'root' })
export class CampaignService {
  private apiUrl = `${environment.apiUrl}/campaigns`;
  
  constructor(private http: HttpClient) { }
  
  // GET request
  getAllCampaigns(): Observable<Campaign[]> {
    return this.http.get<Campaign[]>(this.apiUrl);
  }
  
  // POST request
  createCampaign(campaign: CampaignRequest): Observable<Campaign> {
    return this.http.post<Campaign>(this.apiUrl, campaign);
  }
  
  // PUT request
  updateCampaign(id: number, campaign: CampaignRequest): Observable<Campaign> {
    return this.http.put<Campaign>(`${this.apiUrl}/${id}`, campaign);
  }
  
  // DELETE request
  deleteCampaign(id: number): Observable<void> {
    return this.http.delete<void>(`${this.apiUrl}/${id}`);
  }
}
```

# 8️⃣ ROUTING & NAVIGATION

## Routing Concept

Routing determines **which component to display based on the URL**.

```
URL: /login              → Display LoginComponent
URL: /campaigns          → Display CampaignListComponent
URL: /dashboard/admin    → Display AdminDashboardComponent
URL: /campaigns/123      → Display CampaignDetailComponent
```

## Complete Route Configuration

```typescript
// app-routing.module.ts
const routes: Routes = [
  // Home redirect
  { path: '', redirectTo: '/login', pathMatch: 'full' },
  
  // Auth routes (no protection)
  { path: 'login', component: LoginComponent },
  { path: 'signup', component: SignupComponent },
  
  // Admin dashboard (protected + role-based)
  {
    path: 'dashboard/admin',
    component: AdminDashboardComponent,
    canActivate: [AuthGuard, RoleGuard],
    data: { role: 'ADMIN' }  // Pass role requirement
  },
  
  // Brand dashboard  
  {
    path: 'dashboard/brand',
    component: BrandDashboardComponent,
    canActivate: [AuthGuard, RoleGuard],
    data: { role: 'BRAND' }
  },
  
  // Influencer dashboard
  {
    path: 'dashboard/influencer',
    component: InfluencerDashboardComponent,
    canActivate: [AuthGuard, RoleGuard],
    data: { role: 'INFLUENCER' }
  },
  
  // Campaign routes
  {
    path: 'campaigns',
    component: CampaignListComponent,
    canActivate: [AuthGuard]  // Just login required
  },
  {
    path: 'campaigns/new',
    component: CampaignFormComponent,
    canActivate: [AuthGuard, RoleGuard],
    data: { role: 'BRAND' }  // Only brands can create
  },
  {
    path: 'campaigns/edit/:id',  // :id is a route parameter
    component: CampaignFormComponent,
    canActivate: [AuthGuard, RoleGuard],
    data: { role: 'BRAND' }
  },
  {
    path: 'campaigns/:id',  // View campaign details
    component: CampaignDetailComponent,
    canActivate: [AuthGuard]
  },
  
  // Other feature routes
  { path: 'sponsorship-requests', component: SponsorshipRequestComponent, canActivate: [AuthGuard] },
  { path: 'payments', component: PaymentComponent, canActivate: [AuthGuard] },
  { path: 'ratings', component: RatingComponent, canActivate: [AuthGuard] },
  { path: 'notifications', component: NotificationComponent, canActivate: [AuthGuard] },
  
  // Wildcard (404) - must be last
  { path: '**', redirectTo: '/login' }
];

@NgModule({
  imports: [RouterModule.forRoot(routes)],
  exports: [RouterModule]
})
export class AppRoutingModule { }
```

## Programmatic Navigation

# 9️⃣ ROUTE GUARDS - Protecting Routes

## What are Route Guards?

Route guards **control access to routes**. They answer: "Should user access this route?"

## AuthGuard - Is User Logged In?

```typescript
// guards/auth.guard.ts
import { Injectable } from '@angular/core';
import { CanActivate, Router } from '@angular/router';
import { AuthService } from '../services/auth.service';

@Injectable({
  providedIn: 'root'
})
export class AuthGuard implements CanActivate {
  
  constructor(private authService: AuthService, private router: Router) { }
  
  canActivate(): boolean {
    // Check if user is logged in
    if (this.authService.isLoggedIn()) {
      return true;  // Allow access
    }
    
    // Not logged in, redirect to login
    this.router.navigate(['/login']);
    return false;  // Deny access
  }
}

// Usage in routes
{ path: 'campaigns', component: CampaignListComponent, canActivate: [AuthGuard] }
```

## RoleGuard - Does User Have Required Role?

```typescript
// guards/role.guard.ts
import { Injectable } from '@angular/core';
import { CanActivate, Router, ActivatedRouteSnapshot } from '@angular/router';
import { AuthService } from '../services/auth.service';

@Injectable({
  providedIn: 'root'
})
export class RoleGuard implements CanActivate {
  
  constructor(private authService: AuthService, private router: Router) { }
  
  canActivate(route: ActivatedRouteSnapshot): boolean {
    // Get required role from route data
    const requiredRole = route.data['role'];
    
    // Check if user has the role
    if (this.authService.hasRole(requiredRole)) {
      return true;  // Allow access
    }
    
    // Wrong role, deny access
    this.router.navigate(['/login']);
    return false;
  }
}

// Usage in routes
{
  path: 'dashboard/admin',
  component: AdminDashboardComponent,
  canActivate: [AuthGuard, RoleGuard],
  data: { role: 'ADMIN' }  // Only admins allowed
}
```

## Guard Execution Flow

```
User clicks: /dashboard/admin
          ↓
Execute AuthGuard.canActivate()
          ↓
Is user logged in? → NO → Redirect to /login
          ↓ YES
Execute RoleGuard.canActivate()
          ↓
Does user have 'ADMIN' role? → NO → Redirect to /login
          ↓ YES
Navigate to AdminDashboardComponent
```

# 🔟 HTTP INTERCEPTORS - Middleware for API Calls

## What is an Interceptor?

An interceptor **intercepts HTTP requests & responses** to add custom logic.

```
Before Interceptor:
Component → HTTP Service → Backend

With Interceptor:
Component → HTTP Service → Interceptor (can modify request) → Backend
         ← ← ← ← ← ← ← ← Interceptor (can modify response) ← ← ← 
```

## AuthInterceptor - Add JWT Token

```typescript
// interceptors/auth.interceptor.ts
import { Injectable } from '@angular/core';
import { HttpInterceptor, HttpRequest, HttpHandler, HttpEvent, HttpErrorResponse } from '@angular/common/http';
import { Observable, throwError } from 'rxjs';
import { catchError } from 'rxjs/operators';
import { AuthService } from '../services/auth.service';
import { Router } from '@angular/router';

@Injectable()
export class AuthInterceptor implements HttpInterceptor {
  
  constructor(private authService: AuthService, private router: Router) { }
  
  intercept(request: HttpRequest<any>, next: HttpHandler): Observable<HttpEvent<any>> {
    
    // 1. Get the JWT token from service
    const token = this.authService.getToken();
    
    // 2. If token exists, add it to request header
    if (token) {
      request = request.clone({
        setHeaders: {
          Authorization: `Bearer ${token}`
        }
      });
    }
    
    // 3. Pass request to backend
    // 4. Handle response
    return next.handle(request).pipe(
      catchError((error: HttpErrorResponse) => {
        // 5. Handle 401 (Unauthorized)
        if (error.status === 401) {
          // Token expired, logout user
          this.authService.logout();
          this.router.navigate(['/login']);
        }
        return throwError(() => error);
      })
    );
  }
}

// Register in AppModule
providers: [
  {
    provide: HTTP_INTERCEPTORS,
    useClass: AuthInterceptor,
    multi: true  // Allow multiple interceptors
  }
]
```

## Interceptor Flow

```
Component makes request:
this.http.get('/api/campaigns')
     ↓
AuthInterceptor.intercept()
  1. Gets token from AuthService
  2. Adds: Authorization: Bearer eyJhbGciOiJIUzI1NiI...
  3. Clones request with new header
     ↓
Request sent to backend:
GET /api/campaigns
Headers: Authorization: Bearer eyJhbGciOiJIUzI1NiI...
     ↓
Backend receives request with token
Backend validates token
Backend sends response
     ↓
AuthInterceptor receives response
  If 401: Logout user
  Otherwise: Pass response to component
     ↓
Component receives response data
```

# 1️⃣3️⃣ STATE MANAGEMENT WITH RXJS - Data Flow

## What is RxJS?

RxJS = **Reactive Extensions for JavaScript**

Allows **reactive data flow** (data streams that change over time).

## Observables vs Traditional

```
Traditional (Imperative):
let user = null;
user = getUserFromBackend();  // Single value
console.log(user);            // Use once

RxJS (Reactive):
currentUser$ = new BehaviorSubject(null);
currentUser$.subscribe(user => {
  console.log('User changed to:', user);
});  // Reacts every time user changes
```

## BehaviorSubject - State Container

```typescript
// Service maintains state
@Injectable({ providedIn: 'root' })
export class AuthService {
  // Create a BehaviorSubject (state container)
  private currentUserSubject = new BehaviorSubject<User | null>(null);
  
  // Expose as Observable (read-only)
  public currentUser$ = this.currentUserSubject.asObservable();
  
  // Update state
  login(credentials): Observable<AuthResponse> {
    return this.http.post('/api/auth/login', credentials).pipe(
      tap(response => {
        // When response received, update BehaviorSubject
        this.currentUserSubject.next(response.user);
      })
    );
  }
  
  logout(): void {
    // Update state
    this.currentUserSubject.next(null);
  }
  
  getCurrentUser(): User | null {
    // Get current value
    return this.currentUserSubject.value;
  }
}

// Component subscribes to state
export class NavbarComponent implements OnInit, OnDestroy {
  currentUser: User | null = null;
  private subscription: Subscription;
  
  constructor(private authService: AuthService) { }
  
  ngOnInit() {
    // Subscribe to user changes
    this.subscription = this.authService.currentUser$.subscribe(user => {
      this.currentUser = user;  // Update when state changes
    });
  }
  
  ngOnDestroy() {
    // Cleanup: stop listening
    this.subscription.unsubscribe();
  }
}
```

## Observable Operators

```typescript
// tap() - do something without changing data
this.http.get('/api/campaigns')
  .pipe(
    tap(campaigns => console.log('Loaded:', campaigns))
  )
  .subscribe();

// map() - transform data
this.authService.login(...)
  .pipe(
    map(response => response.user)
  )
  .subscribe(user => console.log(user));

// filter() - only emit if condition met
this.currentUser$
  .pipe(
    filter(user => user !== null),
    tap(user => console.log('User logged in:', user))
  )
  .subscribe();

// switchMap() - chain observables
this.route.params
  .pipe(
    switchMap(params => this.campaignService.getCampaignById(params['id']))
  )
  .subscribe(campaign => console.log(campaign));
```

## Observable vs Promise

```
Promise:
- Single value
- Resolved once
- No cancellation

Observable:
- Multiple values
- Resolved many times
- Can unsubscribe (cancel)
- Better for streams of data
- Can be transformed with operators
```

## What is a Single Page Application (SPA)?

```
Traditional Multi-Page Website:
Click link → Browser requests new HTML → Page reloads → See new page
Result: Blank flash, page blinks

Single Page Application (Angular):
Click link → JavaScript handles it → Only content updates → Smooth transition
Result: Smooth, desktop-like experience

Advantages of SPA:
✓ Fast - No page reload blinks
✓ Smooth - Animated transitions
✓ User-friendly - Like native apps
✓ Less bandwidth - Only data sent, not full HTML
```

## HTML Data Binding Explained

```html
<!-- 1. Interpolation - Show component data -->
<p>Hello {{ username }}</p>
<!-- If username = 'John', shows: Hello John -->

<!-- 2. Property Binding - Set HTML property -->
<img [src]="imageUrl">
<button [disabled]="isLoading">

<!-- 3. Event Binding - Handle user actions -->
<button (click)="handleClick()">Click me</button>
<input (keyup.enter)="handleEnter()">

<!-- 4. Two-Way Binding - Both directions -->
<input [(ngModel)]="username">
<!-- Type in input → username updates -->
<!-- Change username in code → input updates -->

<!-- 5. Conditional Rendering -->
<div *ngIf="isLoggedIn">Welcome back!</div>
<div *ngIf="!isLoggedIn">Please login</div>

<!-- 6. Loops -->
<div *ngFor="let campaign of campaigns">
  <p>{{ campaign.title }}</p>
</div>
```

## How Dependency Injection Works

```
WITHOUT DI (Manual):
- Component creates its own service instance
- If 10 components need AuthService, 10 instances exist
- Lots of code: new AuthService()
- Hard to test: Can't swap in mock

WITH DI (Angular Automatic):
- Angular creates one instance
- All components share the same instance (Singleton)
- Just declare: constructor(private authService: AuthService)
- Easy to test: Can inject mock
```

## What is TypeScript?

```
JavaScript + Types

JavaScript (no types):
let user = { name: 'John', age: 30 };
user.age = 'thirty';  // Works but wrong!

TypeScript (with types):
interface User { name: string; age: number; }
let user: User = { name: 'John', age: 30 };
user.age = 'thirty';  // ERROR! age must be number

Benefits:
✓ Catch errors early (before code runs)
✓ Better IDE support (autocomplete)
✓ Clearer code (see what properties exist)
✓ Easier refactoring (change type, see all breaks)
```

## Login vs Registration

```
REGISTRATION (Create new account):
- User enters: name, email, password, role
- Backend: Creates new user in database
- Backend: Returns JWT token
- Frontend: Stores token, logs them in

LOGIN (Use existing account):
- User enters: email, password
- Backend: Finds user by email
- Backend: Checks if password matches
- Backend: Returns JWT token
- Frontend: Stores token, shows dashboard
```

## What is JWT Token?

```
JWT = JSON Web Token

Like a passport for APIs:
- Proves you're logged in
- Sent with every API request
- Backend verifies it's valid
- Contains user info (encoded, not hidden)

Example token:
eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJzdWIiOiI1IiwibmFtZSI6IkpvaG4iLCJpYXQiOjE2MzY0NTk5NDZ9.abcdefgh...

Three parts (separated by dots):
1. Header - Type (JWT)
2. Payload - Data ({"id":5, "name":"John"})
3. Signature - Security (Backend can verify)

Flow:
1. User logs in
2. Backend creates JWT with user info
3. Frontend stores in localStorage
4. Frontend sends with every request
5. Backend verifies and allows access
```

## What is localStorage?

```
localStorage = Browser's persistent storage

Like a small database in the browser:
- Persists even after browser closes
- Stored locally on computer (secure-ish)
- Accessible only to JavaScript on that domain
- Each domain has its own storage

Usage:
localStorage.setItem('token', 'eyJhbGc...');
const token = localStorage.getItem('token');
localStorage.removeItem('token');
localStorage.clear();  // Remove all

Our app uses it for:
- JWT token (stays logged in after refresh)
- User info (current user data)
```

## What is CORS?

```
CORS = Cross-Origin Resource Sharing

Problem:
- Frontend at http://localhost:4200
- Backend at http://localhost:7070
- Different origins!
- Browser blocks it (security)

Solution:
- Backend allows requests from http://localhost:4200
- Backend adds: Access-Control-Allow-Origin: http://localhost:4200
- Browser sees header, allows request

Backend config (Spring Boot):
@Configuration
public class CorsConfig {
  @Bean
  public CorsConfigurationSource corsConfigurationSource() {
    // Allow frontend to make requests
  }
}
```
