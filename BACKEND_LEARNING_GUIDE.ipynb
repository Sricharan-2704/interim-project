# 5️⃣ Magic of Dependency Injection with @Autowired

## The Problem @Autowired Solves

### WITHOUT Dependency Injection (Manual):
```java
public class AuthController {
    public AuthController() {
        // Have to manually create service
        this.authService = new AuthService();
        // Need all dependencies of AuthService too!
    }
    
    private AuthService authService;
}
```

### WITH Dependency Injection (Easy):
```java
public class AuthController {
    @Autowired  // Spring says: "I'll create and inject this for you!"
    private AuthService authService;
}
```

## What is @Autowired?

```
@Autowired = "Spring, please automatically insert this dependency for me"

Spring's job:
1. Identifies @Service, @Repository, @Controller classes
2. Creates instances (called "beans")
3. Registers them in its container
4. When it sees @Autowired, it injects the right bean
5. No manual "new" keyword needed!
```

## Real Example:

```java
@Service
public class AuthService {
    @Autowired
    private UserRepository userRepository;  // Spring creates this!
    
    @Autowired
    private PasswordEncoder passwordEncoder;  // Spring creates this!
    
    @Autowired
    private JwtTokenProvider tokenProvider;  // Spring creates this!
    
    public void register(RegisterRequest request) {
        // userRepository is already set by Spring!
        userRepository.save(user);  // Just use it!
    }
}

@RestController
public class AuthController {
    @Autowired
    private AuthService authService;  // Spring creates this!
}
```

## The Wiring Diagram:

```
Spring Container (IoC Container)
│
├─ Creates UserRepository bean
├─ Creates PasswordEncoder bean
├─ Creates JwtTokenProvider bean
│
├─ Creates AuthService bean
│   ├─ injects UserRepository ✓
│   ├─ injects PasswordEncoder ✓
│   └─ injects JwtTokenProvider ✓
│
└─ Creates AuthController bean
    └─ injects AuthService ✓
```

# 6️⃣ Architectural Design Patterns

## Pattern 1: MVC (Model-View-Controller)

```
MODEL (Entity, DTO)        
    ↑
    │
CONTROLLER (AuthController) ←→ VIEW (JSON Response)
    │
    ↑
SERVICE (Business Logic)
    │
    ↑
REPOSITORY (Database Access)
```

## Pattern 2: Layered Architecture

```
┌─────────────────────────┐
│  Controller Layer       │ ← Handles HTTP requests
├─────────────────────────┤
│  Service Layer          │ ← Business logic
├─────────────────────────┤
│  Repository Layer       │ ← Database access
├─────────────────────────┤
│  Data Access Layer      │ ← Database
└─────────────────────────┘
```

Benefits:
- **Separation of Concerns** = Each layer has one job
- **Easy Testing** = Can test each layer separately
- **Easy Maintenance** = Change one layer without affecting others
- **Scalability** = Easy to add new features

## Pattern 3: Repository Pattern

```
Before (Bad):
CampaignService
    └─ Contains SQL queries directly
    └─ If database changes, update all files!

After (Good):
CampaignService
    └─ doesn't know about database
CampaignRepository
    └─ knows HOW to access database
    └─ if database changes, update only repository!
```

## Pattern 4: Service Locator / Dependency Injection

```
Instead of:
public class Controller {
    private Service service = new Service();  // Tightly coupled
}

Use:
public class Controller {
    @Autowired
    private Service service;  // Loosely coupled, easy to test
}
```

## 📦 PACKAGE 6: `security/` (Authentication)

**Purpose:** Verify that users are who they claim to be

**Simple Explanation:** These classes handle login verification and token generation using JWT (JSON Web Tokens)

### What is JWT?

```
JWT is like a concert ticket:
- User logs in with username/password
- Backend creates a JWT token (like issuing a ticket)
- User sends token with every request (like showing ticket at door)
- Backend verifies token is valid
- User can access protected resources
```

### Files in this package:

| File | Purpose |
|------|---------|
| `JwtTokenProvider.java` | Create and validate JWT tokens |
| `JwtAuthenticationFilter.java` | Check token on every request |
| `CustomUserDetailsService.java` | Load user details from database |

### Example: JwtTokenProvider.java (Simplified)

```java
@Component
public class JwtTokenProvider {
    
    @Value("${app.jwt.secret}")      // Secret key from config
    private String jwtSecret;
    
    @Value("${app.jwt.expiration}")  // How long token lasts
    private long jwtExpiration;
    
    // Generate token after successful login
    public String generateToken(Authentication authentication) {
        UserDetails userDetails = (UserDetails) authentication.getPrincipal();
        Date now = new Date();
        Date expiryDate = new Date(now.getTime() + jwtExpiration);
        
        // Create JWT with username, date signed, expiration date
        return Jwts.builder()
                .subject(userDetails.getUsername())  // Email as subject
                .issuedAt(now)                       // When created
                .expiration(expiryDate)              // When expires
                .signWith(getSigningKey())           // Sign with secret
                .compact();                          // Create final token
    }
    
    // Verify token is valid
    public boolean validateToken(String token) {
        try {
            Jwts.parser()
                .verifyWith(getSigningKey())
                .build()
                .parseSignedClaims(token);  // Check if valid
            return true;
        } catch (Exception e) {
            return false;
        }
    }
    
    // Extract email from token
    public String getUsernameFromToken(String token) {
        Claims claims = Jwts.parser()
                .verifyWith(getSigningKey())
                .build()
                .parseSignedClaims(token)
                .getPayload();
        
        return claims.getSubject();  // Returns the email
    }
}
```

**Key Concepts:**
- JWT tokens contain encoded information
- Tokens are signed with a secret key
- Backend can verify tokens without storing sessions
- Tokens expire after a set time
- Frontend must send token with every request

### Example: AuthController.java

```java
@RestController                        // Tells Spring: "This receives HTTP requests"
@RequestMapping("/api/auth")           // Base URL: http://backend/api/auth
@CrossOrigin(origins = "http://localhost:4200")  // Allow frontend
public class AuthController {
    
    @Autowired
    private AuthService authService;   // Inject the service
    
    // ENDPOINT 1: Register a new user
    @PostMapping("/register")          // POST to /api/auth/register
    public ResponseEntity<?> register(@Valid @RequestBody RegisterRequest request) {
        // 1. Call service to handle registration
        AuthResponse response = authService.register(request);
        
        // 2. Send back response with 200 OK status
        return ResponseEntity.ok(response);
    }
    
    // ENDPOINT 2: Login
    @PostMapping("/login")             // POST to /api/auth/login
    public ResponseEntity<?> login(@Valid @RequestBody LoginRequest request) {
        // 1. Call service to handle login
        AuthResponse response = authService.login(request);
        
        // 2. Send back response
        return ResponseEntity.ok(response);
    }
    
    // ENDPOINT 3: Get current user info
    @GetMapping("/me")                 // GET to /api/auth/me
    public ResponseEntity<?> getCurrentUser() {
        // 1. Call service
        User user = authService.getCurrentUser();
        
        // 2. Send back user data
        return ResponseEntity.ok(user);
    }
}
```

- `@GetMapping` = Receives GET requests (retrieve data)
- `@RequestBody` = Frontend data goes into request body

## 📦 PACKAGE 4: `dto/` (Data Transfer Objects)

**Purpose:** Format data for communication between frontend and backend

**Simple Explanation:** DTOs are like envelopes - you don't send the entire User entity, you send only the data the frontend needs.

### Why DTOs?

```
WRONG WAY (send entire entity):
{
  "id": 1,
  "name": "John",
  "email": "john@example.com",
  "password": "encrypted123456789",     ← NEVER send password!
  "role": "INFLUENCER",
  "bio": "...",
  "profileImage": "..."
}

RIGHT WAY (send DTO):
{
  "token": "jwt_token_here",
  "id": 1,
  "name": "John",
  "email": "john@example.com",
  "role": "INFLUENCER"
}  ← Only what frontend needs!
```

### Files in this package:

| File | Purpose |
|------|---------|
| `AuthResponse.java` | Response sent after login/register |
| `LoginRequest.java` | Data frontend sends for login |
| `RegisterRequest.java` | Data frontend sends for signup |
| `ApiResponse.java` | Generic response for all API calls |
| `CampaignRequest.java` | Data for creating/updating campaign |
| `PaymentRequest.java` | Data for payment submission |
| `RatingRequest.java` | Data for submitting rating |

### Example: AuthResponse.java

```java
@Data  // Generates getters, setters, equals, toString
public class AuthResponse {
    private String token;      // JWT token for authentication
    private String type = "Bearer";
    private Long id;           // User ID
    private String name;       // User name
    private String email;      // User email
    private String role;       // User role (BRAND, INFLUENCER, ADMIN)
    
    // Constructor for easy creation
    public AuthResponse(String token, Long id, String name, 
                        String email, String role) {
        this.token = token;
        this.id = id;
        this.name = name;
        this.email = email;
        this.role = role;
    }
}
```

**Key Concepts:**
- DTOs are "view models" - they show what data is exposed
- You can have different DTOs for different purposes
- Sensitive data (like password) is NEVER in DTOs
- Frontend receives DTOs as JSON

## 📦 PACKAGE 3: `service/` (Business Logic Layer)

**Purpose:** The actual WORK happens here

**Simple Explanation:** Services contain the rules and logic of your business. They use repositories to get data, process it, and send back results.

### Files in this package:

| File | Purpose |
|------|---------|
| `AuthService.java` | Handle registration and login |
| `CampaignService.java` | Create campaigns, manage campaigns |
| `SponsorshipService.java` | Handle sponsorship requests |
| `PaymentService.java` | Process payments |
| `RatingService.java` | Save and retrieve ratings |
| `NotificationService.java` | Send notifications |
| `AdminService.java` | Admin-specific operations |

### Example: AuthService.java (Simplified)

```java
@Service  // Tells Spring: "This contains business logic"
public class AuthService {
    
    @Autowired
    private UserRepository userRepository;  // Can use database
    
    public AuthResponse register(RegisterRequest request) {
        // BUSINESS LOGIC here:
        
        // 1. Get the email and normalize it
        String email = request.getEmail().trim().toLowerCase();
        
        // 2. Check if user already exists
        if (userRepository.existsByEmail(email)) {
            throw new RuntimeException("Email already registered");
        }
        
        // 3. Create new User
        User user = new User();
        user.setEmail(email);
        user.setPassword(encryptPassword(request.getPassword()));
        
        // 4. Save to database
        userRepository.save(user);
        
        // 5. Create JWT token
        String token = generateToken(user);
        
        // 6. Return response
        return new AuthResponse(token, user.getId(), user.getEmail());
    }
}
```

**Key Concepts:**
- `@Service` = This class contains business logic
- `@Autowired` = Spring automatically injects dependencies
- Services call repositories to get/save data
- Services do validation and processing
- Services return DTOs to controllers

```
sponsorship-app-backend/
├── src/
│   ├── main/
│   │   ├── java/
│   │   │   └── com/myapp/sponsorshipapp/     ← YOUR CODE STARTS HERE
│   │   │       ├── controller/               ← Receives HTTP requests
│   │   │       ├── service/                  ← Business logic
│   │   │       ├── repository/               ← Database access
│   │   │       ├── entity/                   ← Database models
│   │   │       ├── dto/                      ← Data Transfer Objects
│   │   │       ├── security/                 ← Authentication & JWT
│   │   │       ├── config/                   ← Configuration
│   │   │       ├── exception/                ← Error handling
│   │   │       └── SponsorshipAppBackendApplication.java  ← Main entry point
│   │   └── resources/
│   │       └── application.properties        ← Settings (port, database, etc.)
│   └── test/                                 ← Tests (we'll skip for now)
├── pom.xml                                   ← Maven dependencies
└── README.md                                 ← Documentation
```

## Inside `src/main/java`

The path `com/myapp/sponsorshipapp` is called a **package**.

**Package naming convention:**
- Reverse domain name style: `com.companyname.appname`
- Like a folder hierarchy that prevents naming conflicts
- Different companies can have similar code without confusion

# 1️⃣ What is a Backend?

Think of a web application like a restaurant:
- **Frontend** = The dining area where customers see the menu and order
- **Backend** = The kitchen where food is prepared

Your backend Java application is like the kitchen:
- It receives orders (HTTP requests from frontend)
- It processes them (business logic)
- It stores information (database)
- It sends results back (response to frontend)

## Key Concept: Client-Server Model

```
FRONTEND (Angular in Browser)
    ↓ Sends HTTP Request (What do you want?)
BACKEND (Java Spring Boot Server) 
    ↓ Does Processing
DATABASE (Stores all data)
    ↓
BACKEND (Get data and process)
    ↓ Sends HTTP Response (Here's your answer)
FRONTEND (Shows data to user)
```
